```python
import pandas as pd
```

```python
# Create a datarame
df = pd.read_csv(r"C:\Users\chibw\Documents\Portfolio\PortfolioProjects\RealEstate\kijijiRentalData2024-05-12.csv")
df
```

```python
# Remove duplicates
df = df.drop_duplicates()
df
```

```python
# Find listings without prices and deleteall rows without prices
null_mask = df["price"].notnull()
print(f" length before: {len(df)}")
print(f" length after: {len(df[null_mask])}")
df=df[null_mask]
```

```python
# Check datatypes of the columns
len(df)
df.dtypes
```

```python
# Cleaning columns by removing brackets and quotation marks

pd.options.display.max_colwidth = 200
Bedrooms = []
Bathrooms = []
Price = []
Appliances = []
Amenities = []
Utilities = []

for index, row in df.iterrows():
    df.loc[index:index,'appliances'] =row['appliances'].strip('[]').replace("'", "")
    df.loc[index:index,'amenities'] =row['amenities'].strip('[]').replace("'", "")
    df.loc[index:index,'price'] =row['price'].split('$')[1].replace("'", "")
    df.loc[index:index,'utilities'] =row['utilities'].strip('[]').replace("'", "")
    df.loc[index:index,'bedrooms'] =row['bedrooms'].split(':' )[1].replace("'", "")
    df.loc[index:index,'bathrooms'] =row['bathrooms'].split(':')[1].replace("'", "")
    df.loc[index:index,'datePosted'] = pd.to_datetime(row['datePosted'])
# print(type(Bedrooms), type(Bathrooms), type(Price), type(Appliances), type(Amenities), type(Utilities))
# df.loc[df.bedrooms]=Bedrooms
# df.loc[df.bathrooms]=Bathrooms
# df.loc[df.price]=Price
# dfdf.loc[df.amenities]=Amenities
# df.loc[df.appliances]=Appliances
# df.loc[df.utilities]=Utilities

df
```

```python
# Sperate address column into street, city and postal when available
# For future visualization and analization


street=[]
city=[]
pstalCode=[]
addCol=[]
i=0
for index, row in df.iterrows():
    i+=1
    flag=False
    
    addCol.append(row['address'])
    splitAdd = row['address'].replace(', Canada', '').split(',')
    
    if len(splitAdd) >= 4:
        flag=True
        
        if len(splitAdd[-1]) <= 7:
            if 'ON' in splitAdd[-1] or 'ontario' in splitAdd[-1].lower():
                pstalCode.append(splitAdd[-1].replace('ON','').replace('Ontario','').replace('ontario',''))
                city.append(splitAdd[-3])
                strt=' '.join(splitAdd[:-2])
                street.append(strt)
            elif 'ON' in splitAdd[-2] or 'ontario' in splitAdd[-2].lower():
                pstalCode.append(splitAdd[-1])
                city.append(splitAdd[-3])
                strt=' '.join(splitAdd[:-2])
                street.append(strt)
            else:
                pstalCode.append(splitAdd[-1])
                city.append(splitAdd[-3])
                street.append(splitAdd[:-2])
        elif 'ON' in splitAdd[-1] or 'ontario' in splitAdd[-1].lower():
            pstalCode.append(splitAdd[len(splitAdd)-1].replace('ON','').replace('Ontario','').replace('ontario',''))
            city.append(splitAdd[-2])
            street.append(' '.join(splitAdd[:-2]))
        else:
            pstalCode.append(splitAdd[3].replace('ON','').replace('Ontario','').replace('ontario',''))
            city.append(splitAdd[1])
            street.append(splitAdd[0])
    elif len(splitAdd) == 3:
        flag=True
        if 'ON' in splitAdd[2] or 'ontario' in splitAdd[2].lower():
            pstalCode.append(splitAdd[2].replace('ON','').replace('Ontario','').replace('ontario',''))
            city.append(splitAdd[1])
            street.append(splitAdd[0])
        elif 'ON' in splitAdd[1] or 'ontario' in splitAdd[1].lower():
            pstalCode.append(splitAdd[2].replace('ON','').replace('Ontario','').replace('ontario',''))
            city.append(splitAdd[0])
            street.append('')
        else:
            pstalCode.append(splitAdd[2])
            city.append(splitAdd[1])
            street.append(splitAdd[0])
    elif len(splitAdd) == 2:
        if 'ON' in splitAdd[1] or 'ontario' in splitAdd[1].lower():
            flag=True
            pstalCode.append(splitAdd[1].replace('ON','').replace('Ontario','').replace('ontario',''))
            city.append(splitAdd[0])
            street.append('')
        elif len(splitAdd[1]) <= 7:
            flag=True
            pstalCode.append(splitAdd[1])
            city.append(splitAdd[0])
            street.append('')
    elif len(splitAdd) == 1:
        if 'ON' in splitAdd or 'ontario' in splitAdd[0].lower():
            flag=True
            pstalCode.append(splitAdd.replace('ON','').replace('Ontario','').replace('ontario','').strip())
            city.append('')
            street.append('')
        elif len(splitAdd) <= 7:
            flag=True
            pstalCode.append(splitAdd)
            city.append('')
            street.append('')
    else:
        flag=True
        pstalCode.append('')
        city.append('')
        street.append('')
    if flag == False:
        pstalCode.append('')
        city.append('')
        street.append('')

        
df.insert(df.columns.get_loc('address')+1, 'street', street)
df.insert(df.columns.get_loc('address')+2, 'city', city)
df.insert(df.columns.get_loc('address')+3, 'postalCode', pstalCode)

df
```
